```python
from ALPR import alpr
from match_license_plate import match_license_plate_to_database
from PIL import Image
import os
```

```python
test_dir = "../../test_images/normal_cars_frames"
os.listdir(test_dir)
print(len(os.listdir(test_dir)))
```

```python
detected_lp = 0
correctly_read = 0
correctly_matched = 0
total = 0
total_characters = 0
correctly_read_characters = 0


for folder in os.listdir(test_dir):
    for image in os.listdir(os.path.join(test_dir, folder)):
        image_path = os.path.join(test_dir, folder, image)
        total += 1
        print(total)
        try:
            lp = alpr(image_path)
            detected_lp += 1
            # match = match_license_plate_to_database(lp, "lp_database.txt")

            if lp == folder:
                correctly_read += 1
            else:
                img = Image.open(image_path)
                img.save("../../test_images/AEB/not_correct/not" + image)

            # if match[0] == folder:
            #     correctly_matched += 1  
        except:
            # Save the image that caused the error
            # img = Image.open(image_path)
            # img.save("../../test_images/AEB/not_detected/not" + image)
            continue


        total_characters += len(folder)
        for i in range(len(folder)):
            if folder[i] == lp[i]:
                correctly_read_characters += 1
```

```python
print(f"Detected license plates: {detected_lp}/{total}")
print(f"Correctly read license plates: {correctly_read}/{detected_lp}")
print(f"Correctly matched license plates: {correctly_matched}/{detected_lp}")
print(f"Accuracy: {correctly_matched/total*100}%")
print(f"Correctly read characters: {correctly_read_characters}/{total_characters}")
print(f"Character accuracy: {correctly_read_characters/total_characters*100}%")
```
