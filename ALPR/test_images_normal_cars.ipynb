```python
from ALPR import alpr
from match_license_plate import match_license_plate_to_database
from PIL import Image
import os
import numpy as np
```

```python
dir = "../../test_images/normal_cars"
print(len(os.listdir(dir)))
```

```python
correct = 0
detected = 0
for filename in os.listdir(dir):
    img_path = os.path.join(dir, filename)
    real_plate = filename.split(".")[0]
    try:
        predicted_plate = alpr(img_path)
        if predicted_plate:
            detected += 1
            print(f"Real: {real_plate}, Predicted: {predicted_plate}")
        if real_plate == predicted_plate:
            correct += 1

    except:
        pass

print(f"Accuracy: {correct/len(os.listdir(dir))}")
print(f"Detection rate: {detected/len(os.listdir(dir))}")

```
