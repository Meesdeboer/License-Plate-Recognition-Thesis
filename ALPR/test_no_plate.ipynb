```python
from ALPR import alpr
from match_license_plate import match_license_plate_to_database
from PIL import Image
import os
```

```python
test_dir = "../../test_images/no_plate"
os.listdir(test_dir)
print(len(os.listdir(test_dir)))
```

```python
detections = 0
count = 0
for file in os.listdir(test_dir):
    image_path = os.path.join(test_dir, file)
    count += 1
    try:
        results = alpr(image_path)
        if results:
            detections += 1

    except:
        continue
    print("Processed: ", count, "Detections: ", detections)
```

```python
print("Total images: ", count, "Detections: ", detections)
```
